```python
import pandas as pd
import altair as alt
import folium
```

```python
df = pd.read_csv('Highest_50.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

```python
graph_count = alt.Chart(df).mark_bar().encode(x=alt.X('name:N', sort='y'), y='count', tooltip=['name', 'count'])
graph_count
```

```python
karte = folium.Map(location=[51,11], zoom_start=3)

folium.TileLayer('stamentoner').add_to(karte)
folium.TileLayer('stamenterrain').add_to(karte)
folium.TileLayer('openstreetmap').add_to(karte)
folium.LayerControl().add_to(karte)

for i, r in df.iterrows():
    folium.Marker(location=(r['lat'], r['lon']), popup=f"<strong>{r['name']}:</strong> Wirkungsort von {r['count']} Personen" , tooltip=r['name']).add_to(karte)
karte
```
